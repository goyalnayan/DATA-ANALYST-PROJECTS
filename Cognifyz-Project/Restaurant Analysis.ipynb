```python
# Convert values in 'Cuisines' column to strings
df['Cuisines'] = df['Cuisines'].astype(str)

# Split the 'Cuisines' column
df['Cuisines'] = df['Cuisines'].str.split(', ')

# Explode the list of cuisines into separate rows
df_exploded = df.explode('Cuisines')

# Count occurrences of each cuisine
cuisine_counts = df_exploded['Cuisines'].value_counts()

# Calculate percentages
total_restaurants = len(df)
cuisine_percentages = (cuisine_counts / total_restaurants) * 100

# Display the percentages for the top cuisines
print(cuisine_percentages.head(3))
```
